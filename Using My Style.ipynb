```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline

plt.style.use('~/Documents/python_projects/style_lib/wright.mplstyle')
```

```python
# Make up data
x = np.linspace(0, 10, 100)
y = x ** 2
z = x ** 2 + 10

# Plot Data
plt.plot(x,y, label = 'x versus y')
plt.plot(x,z, label = 'x versus z')
plt.xlabel('x values')
plt.ylabel('y and z values')
plt.title('Two Curves')
plt.legend(loc = (.68,.12));
```
